```python
"""

 pr3 Descriptive Statistics - Measures of Central Tendency and variability
Perform the following operations on any open source dataset (e.g., data.csv)
1. Provide summary statistics (mean, median, minimum, maximum, standard deviation) for
a dataset (age, income etc.) with numeric variables grouped by one of the qualitative
(categorical) variable. For example, if your categorical variable is age groups and
quantitative variable is income, then provide summary statistics of income grouped by the
age groups. Create a list that contains a numeric value for each response to the categorical
variable.
2. Write a Python program to display some basic statistical details like percentile, mean,
standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris-versicolor’
of iris.csv dataset.
Provide the codes with outputs and explain everything that you do in this step.
"""
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Employee_Salary_Dataset.csv")

df.columns = df.columns.str.strip().str.lower()

print("Employee Salary dataset loaded successfully")
df.head()
```

```python
print("Dataset Information:")
df.info()

print("\nShape of Dataset:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns

categorical_columns = df.select_dtypes(include=["object"]).columns

for col in numeric_columns:
    df[col] = df[col].fillna(df[col].median())

for col in categorical_columns:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values after handling:")
print(df.isnull().sum())
```

```python
group_column = "gender"

numeric_variables = ["age", "experience_years", "salary"]

print("Categorical variable used for grouping:", group_column)

print("\nNumeric variables used:")
print(numeric_variables)
```

```python
summary_statistics = df.groupby(group_column)[numeric_variables].agg([
    "mean",
    "median",
    "min",
    "max",
    "std"
])

summary_statistics
```

```python
salary_list_by_gender = df.groupby("gender")["salary"].apply(list)

print("List of salary values grouped by gender:")
print(salary_list_by_gender)
```

```python
print("Basic Statistics of Salary:")

print("Mean:", df["salary"].mean())
print("Median:", df["salary"].median())
print("Minimum:", df["salary"].min())
print("Maximum:", df["salary"].max())
print("Standard Deviation:", df["salary"].std())
```

```python
iris = pd.read_csv("dsbdal.csv")

iris.columns = iris.columns.str.strip().str.lower()

print("Iris dataset loaded successfully")
iris.head()
```

```python
print("Iris Dataset Information:")
iris.info()

print("\nShape of Iris Dataset:")
print(iris.shape)

print("\nColumn Names:")
print(iris.columns)
```

```python
if "id" in iris.columns:
    iris = iris.drop("id", axis=1)

print("Dataset after removing ID column if present:")
iris.head()
```

```python
print("Species values:")
print(iris["species"].unique())
```

```python
iris_numeric_columns = iris.select_dtypes(include=[np.number]).columns

print("Numeric columns in Iris dataset:")
print(iris_numeric_columns)
```

```python
iris_setosa = iris[iris["species"] == "Iris-setosa"]

iris_versicolor = iris[iris["species"] == "Iris-versicolor"]

iris_virginica = iris[iris["species"] == "Iris-virginica"]
```

```python
print("Statistical details of Iris-setosa:")

iris_setosa[iris_numeric_columns].describe()
```

```python
print("Statistical details of Iris-versicolor:")

iris_versicolor[iris_numeric_columns].describe()
```

```python
print("Statistical details of Iris-virginica:")

iris_virginica[iris_numeric_columns].describe()
```

```python
print("Group-wise statistical details of all species:")

iris.groupby("species")[iris_numeric_columns].describe()
```

```python
print("Mean values grouped by species:")

iris.groupby("species")[iris_numeric_columns].mean()
```

```python
print("Standard deviation grouped by species:")

iris.groupby("species")[iris_numeric_columns].std()
```
