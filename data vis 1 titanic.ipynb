```python
""" 
pr8 Data Visualization I
1. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information
about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to
see if we can find any patterns in the data.
2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger
is distributed by plotting a histogram.

"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic1.csv")

df.columns = df.columns.str.strip().str.lower()

print("Titanic dataset loaded successfully")
df.head()
```

```python
print("Dataset Information:")
df.info()

print("\nShape of Dataset:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
df["age"] = df["age"].fillna(df["age"].median())

df["fare"] = df["fare"].fillna(df["fare"].median())

df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])

print("Missing values after handling:")
print(df.isnull().sum())
```

```python
sns.countplot(x="survived", data=df)

plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
sns.countplot(x="sex", hue="survived", data=df)

plt.title("Survival Count Based on Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
sns.countplot(x="class", hue="survived", data=df)

plt.title("Survival Count Based on Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
sns.histplot(df["age"], bins=30, kde=True)

plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
sns.histplot(data=df, x="age", hue="survived", bins=30, kde=True)

plt.title("Age Distribution Based on Survival")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
sns.boxplot(x="class", y="fare", data=df)

plt.title("Fare Distribution Based on Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

plt.show()
```

```python
sns.countplot(x="embarked", hue="survived", data=df)

plt.title("Survival Count Based on Embarked Location")
plt.xlabel("Embarked Location")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
sns.countplot(x="who", hue="survived", data=df)

plt.title("Survival Count Based on Passenger Type")
plt.xlabel("Passenger Type")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
sns.countplot(x="alone", hue="survived", data=df)

plt.title("Survival Count Based on Alone or Not")
plt.xlabel("Alone")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
plt.hist(df["fare"], bins=30)

plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
sns.histplot(df["fare"], bins=30, kde=True)

plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")

plt.show()
```
