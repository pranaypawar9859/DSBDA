```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("dsbdal.csv")

df.columns = df.columns.str.strip()

print("Dataset loaded successfully")
df.head()
```

```python
print("Dataset Description:")
print("This dataset is loaded from local device CSV file.")
print("It contains multiple variables/columns for data wrangling operations.")
print("Source: Local CSV file from device")

print("\nColumn Names:")
print(df.columns)
```

```python
print("Number of rows and columns:")
print(df.shape)

print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
print("Statistical summary of numeric columns:")
df.describe()
```

```python
print("Description of all columns:")
df.describe(include="all")
```

```python
print("Data types of each column:")
print(df.dtypes)
```

```python
variable_description = pd.DataFrame({
    "Variable": df.columns,
    "Data Type": df.dtypes.values,
    "Missing Values": df.isnull().sum().values,
    "Unique Values": df.nunique().values
})

variable_description
```

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns

categorical_columns = df.select_dtypes(include=["object"]).columns

print("Numeric Columns:")
print(numeric_columns)

print("\nCategorical Columns:")
print(categorical_columns)
```

```python
for col in categorical_columns:
    df[col] = df[col].astype("category")

print("Data types after converting object columns to category:")
print(df.dtypes)
```

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns

scaler = MinMaxScaler()

df_normalized = df.copy()

df_normalized[numeric_columns] = scaler.fit_transform(df[numeric_columns])

print("Normalized data:")
df_normalized.head()
```

```python
df_encoded = df.copy()

label_encoder = LabelEncoder()

for col in categorical_columns:
    df_encoded[col] = label_encoder.fit_transform(df_encoded[col].astype(str))

print("Dataset after Label Encoding:")
df_encoded.head()
```

```python
df_one_hot = pd.get_dummies(df, columns=categorical_columns)

print("Dataset after One-Hot Encoding:")
df_one_hot.head()
```

```python
print("Final Encoded Dataset Information:")
df_encoded.info()

print("\nFinal Dataset Shape:")
print(df_encoded.shape)

print("\nFinal Encoded Dataset:")
df_encoded.head()
```
