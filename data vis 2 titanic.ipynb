```python
"""
pr9 Data Visualization II
1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution
of age with respect to each gender along with the information about whether they survived
or not. (Column names : 'sex' and 'age')
2. Write observations on the inference from the above statistics.
"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic1.csv")

df.columns = df.columns.str.strip().str.lower()

print("Titanic dataset loaded successfully")
df.head()
```

```python
print("Dataset Information:")
df.info()

print("\nShape of Dataset:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
df["age"] = df["age"].fillna(df["age"].median())

print("Missing values after filling age:")
print(df.isnull().sum())
```

```python
print("Gender values:")
print(df["sex"].unique())

print("\nSurvived values:")
print(df["survived"].unique())
```

```python
sns.boxplot(x="sex", y="age", hue="survived", data=df)

plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")

plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.boxplot(x="sex", y="age", hue="survived", data=df)

plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.legend(title="Survived", labels=["Not Survived", "Survived"])

plt.show()
```
