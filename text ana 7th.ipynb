```python
import nltk
import pandas as pd
import math

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
document = """
Text analytics is the process of converting unstructured text data into meaningful information.
Students are learning text preprocessing techniques such as tokenization, stemming, lemmatization,
stop word removal, and POS tagging. These techniques are useful in natural language processing.
"""

print("Sample Document:")
print(document)
```

```python
document_lower = document.lower()

print("Document after converting to lowercase:")
print(document_lower)
```

```python
tokens = word_tokenize(document_lower)

print("Tokens:")
print(tokens)
```

```python
words = []

for token in tokens:
    if token.isalpha():
        words.append(token)

print("Words after removing punctuation:")
print(words)
```

```python
pos_tags = nltk.pos_tag(words)

print("POS Tags:")
print(pos_tags)
```

```python
stop_words = set(stopwords.words("english"))

filtered_words = []

for word in words:
    if word not in stop_words:
        filtered_words.append(word)

print("Words after stop words removal:")
print(filtered_words)
```

```python
stemmer = PorterStemmer()

stemmed_words = []

for word in filtered_words:
    stemmed_words.append(stemmer.stem(word))

print("Stemmed Words:")
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = []

for word in filtered_words:
    lemmatized_words.append(lemmatizer.lemmatize(word))

print("Lemmatized Words:")
print(lemmatized_words)
```

```python
documents = [
    "Text analytics converts text data into meaningful information",
    "Students are learning text preprocessing techniques",
    "Tokenization stemming and lemmatization are important text processing steps"
]

print("Documents:")
for doc in documents:
    print(doc)
```

```python
processed_documents = []

for doc in documents:
    doc = doc.lower()
    
    tokens = word_tokenize(doc)
    
    words = []
    
    for token in tokens:
        if token.isalpha() and token not in stop_words:
            words.append(token)
    
    processed_documents.append(words)

print("Processed Documents:")
print(processed_documents)
```

```python
vocabulary = []

for doc in processed_documents:
    for word in doc:
        if word not in vocabulary:
            vocabulary.append(word)

print("Vocabulary:")
print(vocabulary)
```

```python
tf_values = []

for doc in processed_documents:
    tf_doc = {}
    
    for word in vocabulary:
        tf_doc[word] = doc.count(word) / len(doc)
    
    tf_values.append(tf_doc)

tf_table = pd.DataFrame(tf_values)

print("Term Frequency Table:")
tf_table
```

```python
idf_values = {}

total_documents = len(processed_documents)

for word in vocabulary:
    document_count = 0
    
    for doc in processed_documents:
        if word in doc:
            document_count = document_count + 1
    
    idf_values[word] = math.log(total_documents / document_count)

idf_table = pd.DataFrame([idf_values])

print("Inverse Document Frequency Table:")
idf_table
```

```python
tfidf_values = []

for tf_doc in tf_values:
    tfidf_doc = {}
    
    for word in vocabulary:
        tfidf_doc[word] = tf_doc[word] * idf_values[word]
    
    tfidf_values.append(tfidf_doc)

tfidf_table = pd.DataFrame(tfidf_values)

print("TF-IDF Table:")
tfidf_table
```
